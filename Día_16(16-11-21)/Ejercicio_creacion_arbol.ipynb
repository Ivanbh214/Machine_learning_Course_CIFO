#Ejercicio sobre el arbol de hijos.

```python
class Node:

    def __init__(self, nodeId = 0, pare = None):
        self.nodeId = nodeId
        self.fills = []
        self.pare = pare
    
    def afegeixFills(self, nodeList):
        self.fills.extend(nodeList)
        for node in nodeList: node.pare = self
    
    def afegeixPare(self, nodePare):
        self.pare = nodePare

    def esArrel(self): return self.pare == None

    def esFulla(self): return self.fills == []

    def generacio(self):
        g = 0
        node = self
        while True:
            if node.esArrel(): return g
            node = node.pare
            g += 1
    
    def mostra(self):
        print("Node:", self.nodeId)
        print("Fills:", [fill.nodeId for fill in self.fills])
        if self.pare: print("Pare:", self.pare.nodeId)
        else: print("Pare: Cap")


class Arbre:

    def __init__(self, nodeArrel):
        self.nodeArrel = nodeArrel
        self.dictNodes = None
        self.totsElsNodes()

    def cercaEnProfunditatIds(self, node = None):
        if node == None: node = self.nodeArrel
        ordreCerca = [node.nodeId]
        for nodeFill in node.fills:
            if nodeFill.esFulla(): ordreCerca.append(nodeFill.nodeId)
            else:
                ordreCerca.extend(self.cercaEnProfunditatIds(nodeFill))
        return ordreCerca

    def cercaEnProfunditatNodes(self, node = None):
        if node == None: node = self.nodeArrel
        ordreCerca = [node]
        for nodeFill in node.fills:
            if nodeFill.esFulla(): ordreCerca.append(nodeFill)
            else:
                ordreCerca.extend(self.cercaEnProfunditatNodes(nodeFill))
        return ordreCerca
    
    def totsElsNodes(self):
        dictNodes = {}
        for node in self.cercaEnProfunditatNodes():
            g = node.generacio()
            if g in dictNodes: dictNodes[g].append(node)
            else: dictNodes[g] = [node]
        for g in dictNodes: dictNodes[g].sort(key = lambda x: x.nodeId)
        self.dictNodes = dictNodes

    def cercaEnAmpladaIds(self):
        ordreCerca = []
        for g in sorted(self.dictNodes):
            #ordreCerca.extend(sorted([node.nodeId for node in self.dictNodes[g]]))
            ordreCerca.extend([node.nodeId for node in self.dictNodes[g]])
        return ordreCerca

    def cercaEnAmpladaNodes(self):
        ordreCerca = []
        for g in sorted(self.dictNodes):
            ordreCerca.extend(self.dictNodes[g])
        return ordreCerca
    
    def mostra(self):
        print("-" * 70)
        print("Node arrel:", self.nodeArrel.nodeId, "\n")
        for node in self.cercaEnAmpladaNodes(): node.mostra()
        print("-" * 70)


nodeArrel = Node()

node1 = Node(1)
node2 = Node(2)

nodeArrel.afegeixFills([node1, node2])


node3 = Node(3)
node4 = Node(4)

node5 = Node(5)
node6 = Node(6)


node1.afegeixFills([node3, node4])
node2.afegeixFills([node5, node6])


nodeArrel.mostra()

node1.mostra()
node2.mostra()

node3.mostra()
node4.mostra()

node5.mostra()
node6.mostra()


arbre = Arbre(nodeArrel)

arbre.mostra()

ordreCercaEnProfunditat = arbre.cercaEnProfunditatIds()

print("\nOrdre de la cerca en profunditat:", ordreCercaEnProfunditat, "\n")

ordreCercaEnAmplada = arbre.cercaEnAmpladaIds()

print("\nOrdre de la cerca en amplada:", ordreCercaEnAmplada, "\n")

#print([node.nodeId for node in arbre.cercaEnProfunditatNodes()])
#print([node.nodeId for node in arbre.cercaEnAmpladaNodes()]) 
```
